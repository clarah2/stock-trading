# Conversions

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read data
data = pd.read_csv('data/filtered_with_ti.csv', header=0)
```

```python
all_tickers = data['tic'].unique()
```

```python
# Add a column to all our data
data['multiplier'] = 1
```

```python
data
```

```python
for tic in all_tickers:
    df = data[data['tic'] == tic]
    # Normal stock split
    prices = df[['Date', 'Open', 'Close', 'multiplier']]
    prices['Open'] = prices['Open'].shift(-1)
    prices.loc[prices['Close'] / prices['Open'] > 1.9, 'multiplier'] = (prices.loc[prices['Close'] / prices['Open'] > 1.9, 'Close'] / prices.loc[prices['Close'] / prices['Open'] > 1.9, 'Open']).round()

    # Reverse stock split
    prices.loc[prices['Open'] / prices['Close'] > 1.9, 'multiplier'] = prices.loc[prices['Open'] / prices['Close'] > 1.9, 'Close'] / prices.loc[prices['Open'] / prices['Close'] > 1.9, 'Open']
    
    data.loc[data['tic'] == tic, 'multiplier'] = prices['multiplier']
```
